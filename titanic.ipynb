## Corrélation entre les variables (matrice de corrélation)

```python
matrice_cor = df_encoded.corr()  # retourne la matrice de corrélation 
print(matrice_cor)               # affiche la martrice
```

```python
plt.figure(figsize=(10, 8))
```

```python
sns.heatmap(matrice_cor,            # les données de matrice_cor seront utiliser
            annot=True,             # affiche les valeurs numériques dans chaque case 
            fmt=".2f",              # formate les valeurs avec 2 décimales
            cmap="coolwarm",        # applique un dégradé de couleurs(bleu:+;rouge:-) 
            square=True)            # force la case à être carrée
plt.title("Matrice de corrélation") # ajoute un titre au graphique
plt.show()                          # affiche le graphique à l'écran
```

```python
# Cette matrice permet de détecter des corrélations élevées entre certaines variables 
# utile pour l’ingénierie des features ou la réduction de dimension.
# Ingénérie des features: c’est l’art de préparer, transformer et créer des variables 
# explicatives (features) à partir des données brutes pour améliorer les performances d’un modèle
```

```python
# Créer un masque pour le triangle supérieur
import numpy as np
mask = np.triu(np.ones_like(matrice_cor, dtype=bool)) # crée une matrice triangulaire supérieure
                                                      # (True au-dessus de la diagonale).
plt.figure(figsize=(10, 8))
sns.heatmap(matrice_cor, 
            mask=mask,   # indique à seaborn de masquer ces cellules.
            annot=True, 
            fmt=".2f", 
            cmap="coolwarm", 
            square=True)
plt.title("Matrice de corrélation")
plt.show()
```
